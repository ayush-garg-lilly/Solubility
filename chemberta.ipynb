```python
# Core libraries
import torch  # PyTorch for deep learning
from transformers import AutoTokenizer, AutoModel  # Hugging Face Transformers for ChemBERTa
import pandas as pd  # Data manipulation
import numpy as np  # Numerical operations

# Machine learning libraries
from sklearn.model_selection import train_test_split  # Splitting dataset
from sklearn.metrics import mean_squared_error, r2_score  # Evaluation metrics
from xgboost import XGBRegressor  # Regression model

# Optional visualization libraries
import matplotlib.pyplot as plt
from tqdm import tqdm  # Progress bar
```

```python
from transformers import AutoTokenizer

# Initialize ChemBERTa tokenizer
tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
print("Tokenizer initialized successfully.")
```

```python
df_merged['combined_SMILES'] = df_merged['solute_smiles'] + " | " + df_merged['solvent_smiles']
inputs = tokenizer(df_merged['combined_SMILES'].tolist(), padding=True, truncation=True, return_tensors="pt")
```

```python
df_merged['combined_SMILES']
```

```python
tokenized_inputs = tokenizer(
    df_merged['combined_SMILES'].tolist(),
    padding=True,
    truncation=True,
    return_tensors="pt"
)
```

```python
print("Tokenized inputs:", tokenized_inputs)
```

```python
# Initialize ChemBERTa model
model = AutoModel.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")

# Generate embeddings from tokenized inputs (disable gradient computation for efficiency)
with torch.no_grad():
    outputs = model(**tokenized_inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)  # Mean pooling across tokens

# Convert embeddings to NumPy array for further processing
embeddings_array = embeddings.numpy()
print("Generated embeddings shape:", embeddings_array.shape)
```

```python
#implemeting ChemBERTa without additional descriptors, with only SMILES
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    embeddings_array, df_merged['solubility_g_100g_log'], test_size=0.2, random_state=42
)

# Train a Random Forest Regressor
regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train, y_train)

# Predict on test set and evaluate performance
y_pred = regressor.predict(X_test)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse:.4f}")
```

```python
rmse = mse ** 0.5
print(f"RMSE: {rmse:.4f}")
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
r2 = r2_score(y_test, y_pred)
print(f"R²: {r2:.4f}")
```

```python
def adjusted_r2(r2, n, p):
    return 1 - (1 - r2) * ((n - 1) / (n - p - 1))
```

```python
n = len(y_test)  # Number of samples
p = 1            # Number of predictors (features)
adj_r2 = adjusted_r2(r2, n, p)
```

```python
print(f"Adjusted R²: {adj_r2:.4f}")
```

```python
mae = mean_absolute_error(y_test, y_pred)
print(f"MAE: {mae:.4f}")
```
